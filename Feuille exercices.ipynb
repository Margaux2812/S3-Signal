# Calculs de TZ
  
## (4)
On a $b[k] \leftrightarrow \cal{B}(z)$ avec sa couronne ouverte $\{ z : R_1 < \mid z\mid < R_2 \}$  
On pose $y[k] = b[k] \alpha^k$.  
$$\cal{Y}(z) = \sum_{-\infty}^{\infty} b[k]\alpha^k z^{-k} \\
=  \sum_{-\infty}^{\infty} b[k] \left( \frac{z}{\alpha} \right) ^{-k} \\
= \cal{B}\left( \frac{z}{\alpha} \right)$$
On sait que $\cal{B}(z)$ a une couronne ouverte $\{ z : R_1 < \mid z\mid < R_2 \}$  , donc $\cal{B}\left( \frac{z}{\alpha} \right)$ a une couronne ouverte $\{ z : R_1 < \mid \frac{z}{\alpha}\mid < R_2 \} \Leftrightarrow \{ z : \mid\alpha\mid R_1 < \mid z\mid < \mid\alpha\mid R_2 \}$

### (5)
On a $u[k] = c_1^{k-2} \Upsilon[k-2] + k*c_2^k\Upsilon[k]$.  
On sait également que $\Upsilon[k] = 1$ si $k \ge 0$.  
On écrit donc :
$$
\cal{U}(z) = \sum_{-\infty}^{\infty} c_1^{k-2}*\Upsilon[k-2]*z^{-k} + \sum_{-\infty}^{\infty} k*c_2^k \Upsilon[k]*z^{-k}$$
On remplace dans le premier membre k-2 par k et donc -k par -k-2
$$
\cal{U}(z) =  \sum_{-\infty}^{\infty} c_1^{k}*\Upsilon[k]*z^{-k-2} + \sum_{-\infty}^{\infty} k*c_2^k \Upsilon[k]*z^{-k}
$$
On sait ensuite que sur tout $]-infty; 0[$,  $\Upsilon[k] = 0$, donc on change nos bornes de la somme et on enlève les  $\Upsilon[k]$.
$$
\cal{U}(z) =  \sum_{0}^{\infty} c_1^{k}z^{-k-2} + \sum_{0}^{\infty} k*c_2^k*z^{-k}
$$
On sépare ensuite le $z^{-k-2}$ en $z^{-k} * z^{-2}$ et ainsi sortir $z^{-2}$ de la somme, puisqu'il ne dépend pas de k. On regroupe aussi $c_1$ avec le $z$, et on se retrouve comme dans la partie (4). Dans le second membre, on regroupe également $c_2$ avec le $z$
$$
\cal{U}(z) =  z^{-2} * \sum_{0}^{\infty} \left(\frac{z}{c_1}\right)^{-k} + \sum_{0}^{\infty} k*\left(\frac{z}{c_2}\right)^{-k}
$$

##### Nous allons nous intéresser au premier membre :
  
On sait que $\sum_{0}^{\infty} a^{k} = \frac{1}{1-a}$ donc $\sum_{0}^{\infty} a^{-k} = \frac{1}{1-\frac{1}{a}}$.  
On applique donc ça a notre cas, et on obtient pour le premier membre :
$$
z^{-2} * \frac{1}{1-\frac{c_1}{z}}
$$
  
##### Maintenant passons au deuxième membre :
  
On remarque que la dérivée de $\left(\frac{z}{c_2}\right)^{-k}$ est $-k * c_2* \left(\frac{z}{c_2}\right)^{k-1}$.  
Ca se ressemble donc beaucoup, il manque juste une multiplication par -1 (car on a k tout court), par $c_2$ et du coup il faut ajouter une puissance de $\frac{z}{c_2}$.  
Le second membre commence donc par $-\frac{z}{c_2} * c_2 = -z$.  
  
On a donc pris la dérivée de $\left(\frac{z}{c_2}\right)^k$, qui aurait donné avec la somme, $\frac{1}{1-\frac{z}{c_2}}$. On dérive cette fraction et on a ainsi on obtient $-\frac{1}{c_2 \left(1 - \frac{z}{c_2}\right)^2}$  
  
##### On réassemble donc nos deux membres et on obtient :
  
$$
\cal{U}(z) =  z^{-2} * \frac{1}{1-\frac{c_1}{z}} - z*\frac{-1}{c_2\left(1-\frac{z}{c_2}\right)^2} \\
= z^{-2} * \frac{1}{1-\frac{c_1}{z}} + \frac{z}{c_2\left(1-\frac{z}{c_2}\right)^2}
$$

## (6)
On a $v[k] = (-1)^k \Upsilon[k]$. On calcule donc sa TZ.  
$$
\cal{V}(z) = \sum_{-\infty}^{\infty} (-1)^k \Upsilon[k] z^{-k}
$$
On enlève les $\Upsilon[k]$ et on remet la somme sur $[0; +\infty[$. Donc on obtient
$$
\cal{V}(z) = \sum_{0}^{\infty} (-1)^k z^{-k}\\
= \sum_{0}^{\infty} (-z)^{-k} \\
= \frac{1}{1+\frac{1}{z}}
$$
  
    
On a $w[k] = (-1)^k (k+1)\Upsilon[k]$. On calcule donc sa TZ.  
$$
\cal{W}(z) = \sum_{-\infty}^{\infty} (-1)^k (k+1)\Upsilon[k] z^{-k}
$$
On enlève les $\Upsilon[k]$ et on remet la somme sur $[0; +\infty[$. Donc on obtient
$$
\cal{W}(z) = \sum_{0}^{\infty} (-1)^k (k+1)z^{-k}\\
= \sum_{0}^{\infty} (k+1)(-\frac{1}{z})^{k}
$$
On voit que ça ressemble à la dérivée de $-\frac{1}{z}^{k+1}$ qui est $(k+1)*z^{-2}*\left(-\frac{1}{z}\right)^k$.  
On va devoir compenser donc avec un $z^2$.  
Donc on a :
$$
\cal{W}(z) = z^2 * \sum_{0}^{\infty} \frac{\partial}{\partial z} \left(-\frac{1}{z}\right)^{k+1}
$$
On sait que la dérivée de $\frac{1}{1+\frac{1}{z}}$ est $\frac{1}{z^2(1+\frac{1}{z})^2}$ donc on a
$$
\cal{W}(z) = z^2 * \frac{1}{z^2(1+\frac{1}{z})^2} \\
= \frac{1}{\left(1+\frac{1}{z}\right)^2}
$$  
  
  
On a $y[k] = k^2\Upsilon[k]$. On calcule donc sa TZ.
$$
\cal{Y}(z) = \sum_{-\infty}^{\infty} k^2\Upsilon[k] z^{-k}\\
= \sum_{0}^{\infty} k^2 z^{-k}
$$
On a la dérivée de $z^{-k}$ qui est $-kz^{-k-1}$. Pour obtenir du $k^2$ il serait donc bien de dériver deux fois, et d'avoir ainsi du $k*k$. On devra compenser avec du $z^2$ comme on perdra deux puissances de $z$. On a donc :
$$
\cal{Y}(z) = -z \sum_{0}^{\infty} \frac{\partial}{\partial z} \left( -z \sum_{0}^{\infty} \frac{\partial}{\partial z} (z^{-k})\right)
$$
On va donc poser $f(z) = \frac{1}{1-\frac{1}{z}}$. On a donc :
$$
\left(-z*f(z)\right)' = \frac{(z-1)^2 -2*z*(z-1)}{(z-1)^4} \\
= \frac{-z^2+1}{(z-1)^4} \\
= \frac{(1-z)(1+z)}{(1-z)^4} \\
= \frac{1+z}{(z-1)^3}
$$
On cherche $f''(z)$ donc on dérive $f'(z)$ :
$$
f''(z) = -(-2)(z-1)^{-3} = \frac{2}{(z-1)^3}
$$
Donc
$$
\cal{Y}(z) = z^2 * \frac{2}{(z-1)^3} = \frac{2z^2}{(z-1)^3}
$$
